```python
#  PRIME NUMBERS
import math
def is_prime(n: int) -> bool:
    """Return True if n is a prime number, else False."""
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limit = int(math.isqrt(n))
    for d in range(3, limit + 1, 2):
        if n % d == 0:
            return False
    return True
num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a PRIME number.")
else:
    print(num, "is NOT a prime number.")
```

```python
# Exercise 4: Word Counter

input_text = "THERE IS AN WATER WHERE I CANT DO ANY THING IN THAT TO PLAY AND THE WATER IS AN "
words = input_text.split()
word_count = {}
for w in words:
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

```python
# Exercise 5: Palindrome Checker

import re

def is_palindrome(s: str) -> bool:
    """Return True if s is a palindrome (ignores case, spaces, punctuation)."""
    cleaned = re.sub(r'[^A-Za-z0-9]', '', s).lower()
    return cleaned == cleaned[::-1]

# Example Tests
tests = ["racecar", "A man, a plan, a canal: Panama!", "hello", "No lemon, no melon"]
for t in tests:
    print(t, "=>", is_palindrome(t))

# User Input
user_text = input("Enter text: ")
print("Palindrome?", is_palindrome(user_text))
```
